### Chart - 1
####*Question no. 1  Monthly ADR trend (City vs Resort Hotels)**

````python
# Monthly ADR Trend by Hotel Type
monthly_adr = df1.groupby(['hotel','arrival_date_month'])['adr'].mean().unstack()
monthly_adr.T.plot(figsize=(12,6), marker='o')
plt.title('Monthly ADR Trend (City vs Resort Hotels)', fontsize=14)
plt.ylabel('Average Daily Rate (ADR)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
````

#### 1. Why did you pick the specific chart?
##### **A line chart is best for showing trends over time and for comparing multiple categories (City vs Resort hotels).**

####2. What is/are the insight(s) found from the chart?
##### **ADR for City hotels is usually more stable, while Resort hotels tend to show higher ADR in peak summer months.**


#### 3. Will the gained insights help creating a positive business impact?
#### Are there any insights that lead to negative growth? Justify with specific reason.
#####Positive: Helps in dynamic pricing strategy  City hotels can maintain consistency, while Resorts can maximize revenue in peak months.

#####Negative: Strong dependence on summer demand for Resorts makes them vulnerable in off-season.

### Chart - 2
####*Question no. 2 Holidays (December, August)  Spikes in Bookings & ADR*

````python
# Bookings & ADR for Holiday Months
holiday_data = df1[df1['arrival_date_month'].isin(['December','August'])]
holiday_summary = holiday_data.groupby('arrival_date_month').agg(
    bookings=('hotel','count'),
    avg_adr=('adr','mean')
)
holiday_summary.plot(kind='bar', secondary_y='avg_adr', figsize=(10,6))
plt.title('Holiday Season: Bookings & ADR')
plt.show()
````

##### 1. Why did you pick the specific chart?

**A bar chart with dual axes shows both volume (bookings) and value (ADR) clearly.**

##### 2. What is/are the insight(s) found from the chart?

**August shows the highest spike in both bookings and ADR (holiday + summer vacation season).**

**December shows moderate bookings but higher ADR (festive premium pricing).**

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive: Hotels can increase holiday packages, promotions, and upselling during these months.

Negative: Heavy reliance on August could cause instability if demand drops due to external factors (e.g., travel restrictions).

### Chart - 4
####*Question no. 4Best Performing Quarters (Occupancy)*

````python
# Convert month name to number
df1['arrival_date_month_num'] = pd.to_datetime(df1['arrival_date_month'], format='%B').dt.month

# Create a full datetime column
df1['arrival_date'] = pd.to_datetime(
    dict(year=df1['arrival_date_year'],
         month=df1['arrival_date_month_num'],
         day=df1['arrival_date_day_of_month'])
)
````

````python
# Add quarter column
df1['quarter'] = df1['arrival_date'].dt.to_period('Q')

# Occupancy rate = 1 - cancellation rate
quarterly_occupancy = df1.groupby('quarter')['is_canceled'].apply(lambda x: 1 - x.mean())

# Plot
quarterly_occupancy.plot(kind='bar', figsize=(10,6))
plt.title('Quarterly Occupancy Performance')
plt.ylabel('Occupancy Rate')
plt.show()
````

##### 1. Why did you pick the specific chart?

**A bar chart clearly compares occupancy rates across Q1Q4.**

##### 2. What is/are the insight(s) found from the chart?

**Q3 (JulySept) performs best due to holiday season.**

**Q1 (JanMar) shows the weakest occupancy.**

##### 3. Will the gained insights help creating a positive business impact?
**Are there any insights that lead to negative growth? Justify with specific reason.**

##### **Positive: Hotels can plan maintenance and cost control in Q1 (low demand) and maximize promotions in Q3.**


**Negative: Strong seasonal reliance reduces year-round stability.**

### Chart - 5
####*Question no. 5 Last-minute bookings by season*

````python
# Lead time distribution by season
df1['arrival_month'] = pd.to_datetime(df1['arrival_date']).dt.month
season_map = {12:'Winter',1:'Winter',2:'Winter',
              3:'Spring',4:'Spring',5:'Spring',
              6:'Summer',7:'Summer',8:'Summer',
              9:'Fall',10:'Fall',11:'Fall'}
df1['season'] = df1['arrival_month'].map(season_map)

season_lead = df1.groupby('season')['lead_time'].median()
season_lead.plot(kind='bar', figsize=(10,6))
plt.title('Median Lead Time by Season')
plt.ylabel('Days Before Booking')
plt.show()
````

##### 1. Why did you pick the specific chart?

**A bar chart of median lead times highlights how far in advance bookings happen across seasons.**

# **Conclusion**

1. Monthly ADR Trend (City vs Resort Hotels):
Resort Hotels show strong seasonality with higher ADR in summer (JulyAugust), while City Hotels remain more stable. This highlights the importance of dynamic pricing, with Resorts capitalizing on holiday demand and City Hotels maintaining year-round competitiveness.

2. Holiday Months (December & August):
Both bookings and ADR peak in August (vacation season), while December shows high ADR due to festive pricing despite moderate bookings. Targeted holiday promotions and premium festive packages can maximize revenue.

3. Seasonality in Cancellations:
Cancellation rates spike in July and August, reflecting high volatility in summer travel. Hotels should apply stricter cancellation policies or flexible pricing in these months to minimize revenue loss.

4. Quarterly Occupancy Performance:
Q3 (JulySeptember) performs best in occupancy, driven by summer holidays, while Q1 (JanuaryMarch) is weakest. Hotels can optimize by focusing on promotions and events in Q1 and maximizing revenue in Q3.

5. Last-Minute Bookings by Season:
Summer attracts more last-minute bookings, whereas Winter shows more planned, early bookings. Hotels can adopt flash deals in Summer and advance booking discounts in Winter to align with guest behavior.

6. Preferred Room Type Allocation:
During peak months (August, December), a higher share of customers do not get their preferred room type. This signals the need for improved room inventory management to reduce dissatisfaction and negative reviews.

7. Meal Type vs ADR:
Guests opting for Half Board (HB) and Full Board (FB) generate higher ADR compared to Bed & Breakfast (BB). Promoting bundled meal packages can increase per-guest revenue.

8. Direct Bookings vs Preferred Rooms:
Guests booking through Direct channels are more likely to receive their preferred room type, reinforcing the importance of strengthening direct booking incentives (loyalty points, discounts).

9. Special Requests by Customer Type:
Transient customers place the highest number of special requests, while Group bookings place the fewest. This insight helps optimize staffing and personalize services where they matter most.

10. Group Bookings by Hotel Type:
City Hotels attract more group bookings (corporate/MICE events), while Resort Hotels cater more to leisure travelers. This underlines the need for City Hotels to strengthen business partnerships and Resorts to focus on seasonal leisure packages.
